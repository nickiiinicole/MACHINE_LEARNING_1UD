## 1. Consulta de datasets precargados
Existen datasets  dentro de la librería scikit-learn que podemos cargar de forma simple para hacer pruebas

````python
from sklearn import datasets

iris_data = datasets.load_iris()
iris_data
````

````python
iris = pd.DataFrame(iris_data.data,columns=iris_data.feature_names)
 # añade una nueva columna llamada "species" aqui guarda el target, 
iris['species'] = pd.Series(iris_data.target_names[iris_data.target])
''' con .head ,muestra los primeros elementos si no pones nada son 5 y si le metes un
parametro le muestra esa cant.'''
iris.head()
````

````python
# Gráfico de puntos
'''muestra como se relacionan el largo y el ancho de los petalos
en las flores del dataset, cada punto seria una flor 
'''
iris.plot.scatter(x='petal length (cm)',
                  y='petal width (cm)')
````

## *Ejercicio 1*

Consulta la librería wine. ¿De qué trata? ¿Qué columnas tiene? Muestra un gráfico de puntos de dos features relevantes.

Wine DataSet , contiene la info. de los componentes del vino, que se elaboran con tres tipos de uva, cuyo objetivo es clasificar las uvas. Esta contiene 13 columnas :

````python
# sklearn -> donde estas los datasets
import pandas as pd
from sklearn.datasets import load_wine

wine_data = load_wine() #cargar datos
# convertir a dataframe -> tabla de datos que se crea con la libreria de panda 
wine = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)
wine['target'] = pd.Series(wine_data.target_names[wine_data.target])

# Ver primeras filas
wine.head()

````
