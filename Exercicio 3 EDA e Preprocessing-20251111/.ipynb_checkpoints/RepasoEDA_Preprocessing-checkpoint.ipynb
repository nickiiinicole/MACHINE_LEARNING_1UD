### *Ejercicio 6: Gráficos*

1. Para cada columna (o feature) numérica, crea un gráfico de barras que muestre su distribución de valores.


```python
import matplotlib.pyplot as plt

numeric_cols = adults_df.select_dtypes(include='number')

for column in numeric_cols.columns:
    plt.figure(figsize=(8,4))
    plt.title(f"Distribución de valores en {column}", fontsize=10, fontstyle='italic', color='purple')
    
    value_counts = numeric_cols[column].value_counts()
    
    plt.bar(value_counts.index, value_counts.values, color='purple', edgecolor='k', alpha=0.7)
    plt.xlabel(column)
    plt.ylabel('Frecuencia')
    plt.grid(True, alpha=0.3)
    plt.show()


```

2. EXTRA: A continuación tienes un código que crea gráficos que muestran la distribución de las ganancias respecto al resto de columnas. Extrae conclusiones de los resultados y propón realizar estos gráficos para otra columna que consideres de interés.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df_numerics = adults_df.select_dtypes(include = np.number)

figure=plt.figure(figsize = (15, 6))

for i, column in enumerate(df_numerics.columns, 1):
    axes = figure.add_subplot(3,3,i)
    sns.kdeplot(x = df_numerics[column], hue = adults_df['gains'], fill = True, ax = axes)
    figure.tight_layout()
```

### *Ejercicio 7: Valores missing*

1. Gestiona los valores nulos con la técnica del rellenado. ¿Qué valores piensas que deberían de tener los nulos presentes en el dataset?

```python
adults_df.isnull().sum()
```

```python
#se cada ve que que uieto nullos pierdo las cosas al borrar si no aporta info si o borramos , borrrar el mala idea ,e tnces solo queda rellenar los valores
# 1con que los relleenamos apra que afecta poco al resultado orginial es meter la media de esta manera influirar poco 
# 2dependiendo del dato que sea cubrir con un valor concreto
# 3mode, el valor que mas se repite a veces es mejor que la media, si hay valores extremos no es ideal 
# 4aleatoriamente  dependendiendo del que contexto
# 5Meter con el valor anterior o posterior


# 1. Rellenar numéricos con mediana
num_cols = adults_df.select_dtypes(include='number').columns
adults_df[num_cols] = adults_df[num_cols].fillna(adults_df[num_cols].median())

# 2. Rellenar categóricos con 'Unknown'
cat_cols = adults_df.select_dtypes(include='object').columns
adults_df[cat_cols] = adults_df[cat_cols].fillna('Unknown')
```
