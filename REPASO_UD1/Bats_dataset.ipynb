## Sección 2: Limpieza y Transformación (Cadenas y Conversión)
Ejercicio 2.1: Conversión de Tipo y Limpieza
La columna Average Lifespan es de tipo object (texto), pero probablemente contiene números y unidades que impiden su uso. Supón que los valores son números seguidos de la palabra "years" (ej. "15 years").

Crea una nueva columna llamada Lifespan_Years.

Limpia la columna original Average Lifespan eliminando el texto " years" (incluyendo el espacio) y convierte los valores resultantes a tipo numérico (float) en la nueva columna Lifespan_Years.

Muestra la media aritmética de la nueva columna Lifespan_Years para verificar la conversión.

```python
average_lifespan_sin_str = bats_df['Average Lifespan'].str.replace(' years', '', regex=False)
# como es 15- 20 years , years ya lo quitamos ahora hay que elegir los rangos y de esos tomar un valor medio
limites = average_lifespan_sin_str.str.split('-', expand=True) #expand true para conseguircolumnas separadas
# ahora calculamos el punto medio 
bats_df['Lifespan_Years']=(limites[0].astype(float)+ limites[1].astype(float)/2)
#anotacion es astype(float) es un metodo de pandas que se aplica a una serie , es mas eficiente(vectorizado), optimizado para operar sobre greandes copnjuntos de datos de forma nativa en c/numpy lo que lo hace mas rapido
bats_df['Lifespan_Years']

bats_df['Lifespan_Years'].mean()
```

Queremos saber para cada Species:

¿Cuál es el Environment que más se repite para esa especie?

```python
bats_df.groupby(by='Species').size().unique
# RECORDAD QUE EN LA MODA , PUEDE DEVOLVER MAS DE UN VALOR SI HAY EMPATE, entonces por eso haces el iloc 
bats_df.groupby('Species').apply(lambda x: x.mode().iloc[0])
# Solución usando agg() ->>> para múltiples estadísticas
bats_df.groupby('Species').agg(
    Habitats_Unicos=('Habitat', 'nunique'),
    Ambiente_Mas_Comun=('Environment', lambda x: x.mode().iloc[0])
)
```
