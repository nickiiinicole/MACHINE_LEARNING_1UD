## Secci칩n 2: Limpieza y Transformaci칩n (Cadenas y Conversi칩n)
Ejercicio 2.1: Conversi칩n de Tipo y Limpieza
La columna Average Lifespan es de tipo object (texto), pero probablemente contiene n칰meros y unidades que impiden su uso. Sup칩n que los valores son n칰meros seguidos de la palabra "years" (ej. "15 years").

Crea una nueva columna llamada Lifespan_Years.

Limpia la columna original Average Lifespan eliminando el texto " years" (incluyendo el espacio) y convierte los valores resultantes a tipo num칠rico (float) en la nueva columna Lifespan_Years.

Muestra la media aritm칠tica de la nueva columna Lifespan_Years para verificar la conversi칩n.

```python
average_lifespan_sin_str = bats_df['Average Lifespan'].str.replace(' years', '', regex=False)
# como es 15- 20 years , years ya lo quitamos ahora hay que elegir los rangos y de esos tomar un valor medio
limites = average_lifespan_sin_str.str.split('-', expand=True) #expand true para conseguircolumnas separadas
# ahora calculamos el punto medio 
bats_df['Lifespan_Years']=(limites[0].astype(float)+ limites[1].astype(float)/2)
#anotacion es astype(float) es un metodo de pandas que se aplica a una serie , es mas eficiente(vectorizado), optimizado para operar sobre greandes copnjuntos de datos de forma nativa en c/numpy lo que lo hace mas rapido
bats_df['Lifespan_Years']

bats_df['Lifespan_Years'].mean()
```
