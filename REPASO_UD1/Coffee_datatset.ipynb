üßÆ 5. Nuevas columnas y transformaciones

Ejercicio 5:
1Ô∏è‚É£ Crea una nueva columna tip que sea el 10% de money

2Ô∏è‚É£ Crea una columna money_category:

'Low' si money < 3

'Medium' si money est√° entre <5

'High' si money > 5

3Ô∏è‚É£ Crea una columna is_morning con valores booleanos (True si Time_of_Day == 'Morning')

```python
coffee_df['tip']=coffee_df['money']*0.10
coffee_df
```

```python
coffee_df['money_category']= coffee_df['tip'].apply(
    lambda money: 'Low' if money<3 else ('Medium' if money <5 else 'High')
)
coffee_df
```

```python
coffee_df['is_morning']= coffee_df['Time_of_Day']=='Morning'
coffee_df
```

Ejercicio 6:

1Ô∏è‚É£ Calcula la media, mediana, m√°ximo y m√≠nimo de money

2Ô∏è‚É£ Muestra la desviaci√≥n est√°ndar

3Ô∏è‚É£ Muestra qu√© hora del d√≠a (morning, afternoon, evening) tiene la mayor media de gasto

4Ô∏è‚É£ Muestra qu√© tipo de caf√© (coffee_name) tiene el mayor gasto medio

```python
media = coffee_df['money'].mean()
median= coffee_df['money'].median()
max_money= coffee_df['money'].max()
min_money= coffee_df['money'].min()

print(f'''
    La media en money es de{ media:.2f}  
    La mediana en money es de{ median:.2f}  
    EL valor maximo en money es { max_money:.2f}  
    EL valor min en money es { min_money:.2f}  
''')
```
